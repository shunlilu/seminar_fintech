```python
#reade csv file
prosper_data = pd.read_csv("./prosperLoanData.csv")
```

```python
#read first 20 lines of data
prosper_data.head(20)
```

```python
#reformat column names
prosper_data.columns = prosper_data.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
#drop rows with CreditGrade:data with CreditGrade is prior to 2008
prosper_data = prosper_data[prosper_data.CreditGrade.isnull()]
#drop rows with people who borrow as a group
prosper_data = prosper_data[prosper_data.GroupKey.isnull()]
```

```python
#drop columns with mostly NaN or irrelevent to result 
prosper_data = prosper_data.drop(columns=['CreditGrade', 'ClosedDate','TotalProsperLoans','OnTimeProsperPayments',\
                            'ProsperPaymentsLessThanOneMonthLate','ProsperPaymentsOneMonthPlusLate',\
                            'ProsperPrincipalBorrowed','ProsperPrincipalOutstanding','ScorexChangeAtTimeOfListing',\
                            'LoanFirstDefaultedCycleNumber','TotalProsperPaymentsBilled','GroupKey','CurrentlyInGroup',\
                                         'ListingNumber','DateCreditPulled','FirstRecordedCreditLine',\
                                          'LoanKey','ProsperRating_Alpha','ProsperScore',\
                                         'TradesNeverDelinquent_percentage','LoanNumber','LoanOriginationDate','MemberKey',\
                                         'LoanMonthsSinceOrigination','ListingCreationDate','ListingKey','IncomeRange',\
                                         'LP_CustomerPayments','LP_CustomerPrincipalPayments','LP_InterestandFees',\
                                         'LP_ServiceFees','LP_CollectionFees','LP_GrossPrincipalLoss','LP_NetPrincipalLoss',\
                                         'LP_NonPrincipalRecoverypayments','BorrowerAPR','LenderYield',\
                                          'EstimatedEffectiveYield','EstimatedLoss','EstimatedReturn'])
##,'LoanStatus','BorrowerRate','ProsperRating_numeric'
```

```python
#clean the rest of data with NaN
prosper_data = prosper_data.dropna()
```

```python
#take average of credit score range of each borrower
prosper_data['AverageCreditScore'] = prosper_data[['CreditScoreRangeLower', 'CreditScoreRangeUpper']].mean(axis=1)
prosper_data = prosper_data.drop(columns=['CreditScoreRangeLower','CreditScoreRangeUpper'])
```

```python
#remove duplicates
prosper_data = prosper_data.drop_duplicates()
```

```python
def encode_dummy(df,name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = f"{name}-{x}"
        df[dummy_name] = dummies[x]
    df.drop(name, axis = 1, inplace = True)
```

```python
encode_dummy(prosper_data, 'Occupation')
encode_dummy(prosper_data, 'BorrowerState')
encode_dummy(prosper_data, 'EmploymentStatus')
encode_dummy(prosper_data, 'LoanOriginationQuarter')
encode_dummy(prosper_data, 'ListingCategory_numeric')
```

```python
prosper_data.loc[prosper_data['LoanStatus'] == 'Current', 'LoanStatus'] = 0
prosper_data.loc[prosper_data['LoanStatus'] == 'Completed', 'LoanStatus'] = 0
prosper_data.loc[prosper_data['LoanStatus'] != 0, 'LoanStatus'] = 1
```

```python
#catogorize the borrwerstate variable
BorrowerState_encoder = LabelEncoder()
BorrowerState_code = BorrowerState_encoder.fit_transform(prosper_data.BorrowerState)
le_BorrowerState_mapping = dict(zip(BorrowerState_encoder.classes_, \
                                    BorrowerState_encoder.transform(BorrowerState_encoder.classes_)))
prosper_data.BorrowerState = BorrowerState_code
le_BorrowerState_mapping
#inverted = label_encoder.inverse_transform(integer_encoded)
```

```python
#catogorize the Occupation variable
Occupation_encoder = LabelEncoder()
Occupation_code = Occupation_encoder.fit_transform(prosper_data.Occupation)
le_Occupation_mapping = dict(zip(Occupation_encoder.classes_, \
                                    Occupation_encoder.transform(Occupation_encoder.classes_)))
prosper_data.Occupation = Occupation_code
le_Occupation_mapping
```

```python
#catogorize the EmploymentStatus variable
EmploymentStatus_encoder = LabelEncoder()
EmploymentStatus_code = EmploymentStatus_encoder.fit_transform(prosper_data.EmploymentStatus)
le_EmploymentStatus_mapping = dict(zip(EmploymentStatus_encoder.classes_, \
                                    EmploymentStatus_encoder.transform(EmploymentStatus_encoder.classes_)))
prosper_data.EmploymentStatus = EmploymentStatus_code
le_EmploymentStatus_mapping
```

```python
#catogorize the LoanOriginationQuarter variable
LoanOriginationQuarter_encoder = LabelEncoder()
LoanOriginationQuarter_code = LoanOriginationQuarter_encoder.fit_transform(prosper_data.LoanOriginationQuarter)
le_LoanOriginationQuarter_mapping = dict(zip(LoanOriginationQuarter_encoder.classes_, \
                                    LoanOriginationQuarter_encoder.transform(LoanOriginationQuarter_encoder.classes_)))
prosper_data.LoanOriginationQuarter = LoanOriginationQuarter_code
le_LoanOriginationQuarter_mapping
```
